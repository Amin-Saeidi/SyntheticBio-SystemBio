```julia
# First Defining Struct and function to save the features we want

mutable struct TheModel
    S
    Metabolites
    Reactions
    Genes
    m
    n
    lb
    ub
    myModel
    reversible_reactions
    irreversible_reactions
end

function method_1_1(model_name, mode=0)
    
    model_file_name = string(model_name,".json")
    download_link = string("http://bigg.ucsd.edu/static/models/", model_file_name)
    
    println("MODEL-NAME: ", model_name)

    !isfile(model_file_name) && download(download_link, model_file_name)
#     myModel = load_model(CoreModel, model_file_name)
#     S = (myModel.S)
    myModel = load_model(StandardModel, model_file_name) # we specifically want to load a StandardModel from the model file
    
  
    n = n_reactions(myModel)
    m = n_metabolites(myModel)
    Metabolites = myModel.metabolites
    Reactions = myModel.reactions
    Genes = myModel.genes
    S = stoichiometry(myModel)
    lb = Dict()
    ub = Dict()
    for i in 1:n_reactions(myModel)
        reaction_id = reactions(myModel)[i]
        lb[reaction_id] = myModel.reactions[reaction_id].lb
        ub[reaction_id] = myModel.reactions[reaction_id].ub    
    end 
    
    model = TheModel(S,Metabolites,Reactions,Genes,m,n,lb,ub,myModel,Dict(),Dict())
    
    return model
            
end
```

```julia
# Function to divide reactions into reversible & irreversible

function method_1_2(model)

    lb = model.lb
    Keys = keys(model.Reactions)
    i = 1  # Fucking Index
    for key in Keys
        reaction_id = key
        if(lb[reaction_id] >= 0)
            model.irreversible_reactions[i] = reaction_id
        else
            model.reversible_reactions[i] = reaction_id
        end
        i = i + 1
    end
    println("Reversible Reactions: ",model.reversible_reactions)
    println("Irreversible Reactions: ",model.irreversible_reactions)
end
```

```julia
# Function to divide reactions into reversible & irreversible

function method_1_3(myModel,re,ire)

    M = 10 ^ 6
    
    for i in keys(ire)
        reaction_id = ire[i]
        myModel.lb[reaction_id] = 0
        myModel.ub[reaction_id]= M
        myModel.myModel.reactions[reaction_id].lb = 0
        myModel.myModel.reactions[reaction_id].ub = M
    end

    for i in keys(re)
        reaction_id = re[i]
        myModel.lb[reaction_id]= -M
        myModel.ub[reaction_id]= M
        myModel.myModel.reactions[reaction_id].lb = -M
        myModel.myModel.reactions[reaction_id].ub = M
    end

end
```

```julia
# Defining the main method
function method_2_1(model)
    
    lb = []
    ub = []
    for i in 1:n_reactions(model.myModel)
        reaction_id = reactions(model.myModel)[i]
        append!(lb, model.myModel.reactions[reaction_id].lb)
        append!(ub, model.myModel.reactions[reaction_id].ub)  
    end
    
    tolerance = 10^(-8)
    
    blocked_reactions = Dict()
    checker = Dict()

    re = 0
    irre = 0
    # For ire : ni
    for i in keys(model.irreversible_reactions)

        # Define the Model
        solver = Model(GLPK.Optimizer)

        # Create Variables
        @variable(solver,lb[i] <= v[i = 1:model.n] <= ub[i])

        # Specify Constraints
        @constraint(solver, (model.S)*v .== 0)
        @constraint(solver, v[i] <= 1)
#         @constraint(solver, v[i] >= lb[i])
#         @constraint(solver, v[i] <= ub[i])
#         @constraint(solver, v .>= lb)
#         @constraint(solver, v .<= ub)
        @objective(solver,Max,v[i])

        # Solving/Optimize
        optimize!(solver)

        if (-tolerance <= objective_value(solver) <= tolerance)
            blocked_reactions[i] = (model.irreversible_reactions[i], "irre")
            irre += 1
        end
    end

    # For re : 2 * nr
    for i in keys(model.reversible_reactions)

        # Define the Model
        solver = Model(GLPK.Optimizer)

        # Create Variables
        @variable(solver,lb[i] <= v[i = 1:model.n] <= ub[i])

        # Specify Constraints
        @constraint(solver, (model.S)*v .== 0)    
        @constraint(solver, v[i] <= 1)
#         @constraint(solver, v[i] >= lb[i])
#         @constraint(solver, v[i] <= ub[i])
#         @constraint(solver, v .>= lb)
#         @constraint(solver, v .<= ub)
        @objective(solver,Max,v[i])

        # Solving/Optimize
        optimize!(solver)

        if (-tolerance <= objective_value(solver) <= tolerance)
            checker[i] = model.reversible_reactions[i]
        end
    end

    for i in keys(model.reversible_reactions)

        # Define the Model
        solver = Model(GLPK.Optimizer)

        # Create Variables
        @variable(solver,lb[i] <= v[i = 1:model.n] <= ub[i])

        # Specify Constraints
        @constraint(solver, (model.S)*v .== 0)    
        @constraint(solver, v[i] >= -1)
#         @constraint(solver, v[i] >= lb[i])
#         @constraint(solver, v[i] <= ub[i])
#         @constraint(solver, v .>= lb)
#         @constraint(solver, v .<= ub)
        @objective(solver,Min,v[i])

        # Solving/Optimize
        optimize!(solver)

        if (-tolerance <= objective_value(solver) <= tolerance) && model.reversible_reactions[i] in values(checker)
            blocked_reactions[i] = (model.reversible_reactions[i], "re")
            re += 1
        end
    end
    println("Number of blocked re reactions: ", re)
    println("Number of blocked irre reactions: ", irre)
    return blocked_reactions
end    
```

```julia
# Defining the main method
function method_2_2(model)
    
    n_irre = length(model.irreversible_reactions)

    one_irre = []

    for i in 1:n_irre
            append!(one_irre, 1)
    end

    lb = []
    ub = []

    for i in 1:n_reactions(model.myModel)
        reaction_id = reactions(model.myModel)[i]
        append!(lb, model.myModel.reactions[reaction_id].lb)
        append!(ub, model.myModel.reactions[reaction_id].ub)  
    end


    # Define the Model
    solver = Model(GLPK.Optimizer)

    # Create Variables
    @variable(solver,lb[i] <= v[i = 1:model.n] <= ub[i])
    @variable(solver,0 <= u[i = 1:n_irre] <= 1)

    # Specify Constraints
    @constraint(solver, (model.S)*v .== 0)
    
    for (index, value) in enumerate(keys(model.irreversible_reactions))
        @constraint(solver, v[value] >= u[index])
    end

    @objective(solver,Max,transpose(one_irre)*u)

    # Solving/Optimize
    optimize!(solver)
    # Objective value
    # @show objective_value(solver)
    
    val = objective_value(solver)
    println("Number of irre Reactions:",n_irre, " --- Number of blocked irre reactions:",n_irre-val )

end    
```
